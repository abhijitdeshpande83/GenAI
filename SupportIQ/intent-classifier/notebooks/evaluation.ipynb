```python
from sklearn.metrics import (accuracy_score, 
                            classification_report,
                            confusion_matrix)
import json
from sklearn.preprocessing import LabelEncoder
```

```python
def get_labels(path):
    with open(path) as f:
        labels = []
        for line in f:
            labels.append(json.loads(line)['label'])
    return labels
```

```python
test_data_dir = '/Volumes/LaCie/Projects_portfolio/NLP/SupportIQ/data/roberta-test.jsonl'
predicted_data_dir = '/Volumes/LaCie/Projects_portfolio/NLP/SupportIQ/output/roberta_predictions_v1.jsonl' 
```

```python
test_labels = get_labels(test_data_dir)
predicted_labels = get_labels(predicted_data_dir)
```

```python
all_labels = list(set(test_labels) | set(predicted_labels))

le = LabelEncoder()
le.fit(all_labels)

y_true = le.transform(test_labels)
y_pred = le.transform(predicted_labels)
```

```python
accuracy_score(y_true, y_pred)
```

```python
print(classification_report(y_true, y_pred))
```
